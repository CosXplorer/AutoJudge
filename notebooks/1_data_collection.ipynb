```python
from datasets import load_dataset
import pandas as pd

# üöÄ streaming = no full download, no disk usage
dataset = load_dataset(
    "open-r1/codeforces",
    split="train",
    streaming=True
)

small = []
for row in dataset:
    small.append(row)
    if len(small) == 5000:   # ‚Üê change this number as you like
        break

df = pd.DataFrame(small)

df.head(), len(df)
```

```python
required_cols = [
    "title",
    "time_limit",
    "memory_limit",
    "description",
    "input_format",
    "output_format",
    "note",
    "rating",
    "tags",
]
```

```python
existing_cols = [c for c in required_cols if c in df.columns]
df_small = df[existing_cols].copy()
```
